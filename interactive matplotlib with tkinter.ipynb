```python
def get_data(rawdata):
    test_data=pd.read_csv(rawdata, skiprows=9)
    test_data.head()
    return test_data
        
```

```python
def plot_data(a):
    
    y_col='INTERNAL'
    
    global xlow, xhigh
    
        
    x=a['Time (seconds)']
    y=a[y_col]
    z=a['BOX TEMP']
    
    y_max=a[y_col].max()
    y_min=a[y_col].min()

    pd.set_option('precision',1)
    plt.style.use('seaborn-whitegrid')

    fig, ax = plt.subplots(4,1,figsize=(15,10))

    xlow=x.min()
    xhigh=x.max()

    ax[0].plot(x,y)
    #         ax[0].set_xlabel(TestData.x_value)
    #         ax[0].set_ylabel(TestData.y_value)
    ax[0].axhline(y=y_max,linewidth=1, color='black',linestyle="--")
    ax[0].axhline(y=y_min,linewidth=1, color='black',linestyle="--")
    ax[0].set_xlim(xlow,xhigh)
    
    n, bins, patches=ax[1].hist(y, 100, density=True,facecolor='g', alpha=0.75)
#     ax[1].plot(bins, y, '--')
    
    
    
    
    ax[2].boxplot(y,showfliers=False,)
    
    
    ax[3].plot(y,z,'ok',markersize=1, alpha=0.2)

    


    plt.show()
```

```python
def get_the_value():
    print(var.get())
    print('jim')
    print(var2.get())
    
    
    subset=a[(a['Time (seconds)']>var.get()) & (a['Time (seconds)']<var2.get())]
    
    
    plot_data(subset)
```
